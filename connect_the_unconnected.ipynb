# Data Cleaning and Preprocessing

```python
import pandas as pd
import numpy as np
df1 = pd.read_csv("Demographic_Data.csv")
df1
```

```python
df2 = pd.read_csv("Hackathon_Call_Log_Data.csv")
df2
```

```python
df2.rename(columns={'REFERRAL_ID': 'REF ID'}, inplace=True)
df2
new_data = pd.merge(df1, df2)
new_data
```

```python
new_data.isna().sum()
```

```python
new_data['HANDLE TIME'].fillna(0, inplace=True)
new_data['TALK TIME'].fillna(0, inplace=True)
new_data['AFTER CALL WORK TIME'].fillna(0, inplace=True)
```

```python
new_data.drop('REFERRAL_TYPE ',  axis=1, inplace=True)
new_data.drop('Referral Date',  axis=1, inplace=True)
new_data.drop('Referral_Code',  axis=1, inplace=True)
new_data.drop('State -  LOB', axis=1, inplace=True)
new_data.dropna(subset=['COUNTY'], inplace=True)
new_data.dropna(subset=['DISPOSITION'], inplace=True)
new_data.dropna(subset=['DAY OF WEEK'], inplace=True)
new_data.drop('LOB',  axis=1, inplace=True)
new_data.drop('REF ID',  axis=1, inplace=True)
new_data.drop('ZIPCODE',  axis=1, inplace=True)
new_data.drop('CALL ID',  axis=1, inplace=True)
```

```python
new_data = new_data[new_data['MEMBER_GENDER'].isin(['M', 'F'])]
```

```python
new_data.isna().sum()
```

```python
new_data['TIMESTAMP'] = new_data['TIMESTAMP'].str.slice(16,)
new_data['TIMESTAMP']
```

```python
#new_data.to_csv('modified.csv')
new_data.head()
```

```python
# Function to define groups based on time intervals
def get_group(hour):
  if 0 <= hour < 6:
    return 'Late Night'
  elif 6 <= hour < 12:
    return 'Morning'
  elif 12 <= hour < 15:
    return 'Afternoon'
  elif 15 <= hour < 18:
    return 'Late noon'
  elif 18 <= hour < 21:
    return 'Evening'
  elif 21 <= hour < 24:
    return 'Night'

# Extract hours for each time column
new_data['hour'] = pd.to_numeric(new_data['TIMESTAMP'].str.split(':', expand=True)[0])

# Apply function to get group labels
new_data['TimeSlot'] = new_data['hour'].apply(get_group)
```

```python
# Function to define groups based on time intervals
def age_group(MEMBER_AGE):
  if 0 <= MEMBER_AGE < 16:
    return 'CHILD'
  elif 16 <= MEMBER_AGE < 30:
    return 'Adults'
  elif 30 <= MEMBER_AGE < 45:
    return 'Youths'
  elif 45 <= MEMBER_AGE < 60:
    return 'Middle-aged'
  elif 60 <= MEMBER_AGE < 80:
    return 'Old-aged'
  else:
    return 'Super-old'


# Apply function to get group labels
new_data['AgeGroup'] = new_data['MEMBER_AGE'].apply(age_group)
```

```python
def calculate_success_rate(group):
  # Filter successful calls (reached or routed based on your outcome columns)
  successful_calls = group[group['DISPOSITION'].isin(['Successful Contact - Assessment',
                                                      'Spoke with Client - Refused Services',
                                                      'Do Not Call','Spoke with Representative',
                                                      'New phone number added in Implify',
                                                      'Interpreter services - Callback Requested',
                                                      'Successful Contact - Visit Schedule',
                                                      'Spoke with Client - Callback Requested','Caller Hangup'])]
  # Calculate success rate as percentage of successful calls divided by total calls
  success_rate = (successful_calls.shape[0] / group.shape[0]) * 100
  group['success_rate']=success_rate
  # Calculate success rate if there are calls in the group
  if group.shape[0] > 0:
    success_rate = (successful_calls.shape[0] / group.shape[0]) * 100
  else:
    success_rate = 0

  # Return aggregated data with success rate
  return pd.DataFrame({'success_rate': [success_rate] * group.shape[0]})
```

```python
# Apply the function to each group and store results in a new Series
success_rates = grouped_data.apply(calculate_success_rate)
```

```python
# Group by time slot and age, calculate success rate, and create a dataframe
success_rates_df = new_data.groupby(['TimeSlot', 'AgeGroup', 'MEMBER_GENDER','DAY OF WEEK']).apply(calculate_success_rate).reset_index()

# Group by age and calculate average success rate
avg_rates_by_age = success_rates_df.groupby('AgeGroup')['success_rate'].mean()

# Sort average success rates by age group
avg_rates_by_age_sorted = avg_rates_by_age.sort_values()

# Plot bar chart with age groups on x-axis and average success rate on y-axis
avg_rates_by_age_sorted.plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Average Success Rate (%)')
plt.title('Average Success Rates by Age')
plt.show()
```

```python
# Group by TimeSlot and calculate the mean success rate for each TimeSlot
avg_success_by_timeslot = success_rates_df.groupby('TimeSlot')['success_rate'].mean()

# Plot the average success rates vs TimeSlot
plt.plot(avg_success_by_timeslot.index, avg_success_by_timeslot.values, marker='o', linestyle='-')
plt.xlabel('Time Slot')
plt.ylabel('Average Success Rate (%)')
plt.title('Average Success Rates vs Time Slot')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if necessary
plt.grid(True)  # Add gridlines for better visualization
plt.show()
```
